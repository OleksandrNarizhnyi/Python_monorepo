```python
# 1. Напишите программу, которая запрашивает у пользователя URL-адрес веб-страницы,
# использует библиотеку Beautiful Soup для парсинга HTML и выводит список всех ссылок
# на странице.

from bs4 import BeautifulSoup
import requests


def html_parsing(url: str) -> list[str]:
    response = requests.get(url).text
    soup_response = BeautifulSoup(response, "html.parser")
    teg = "a"
    all_a = soup_response.find_all(teg)
    all_href = []
    for a in all_a:
        all_href.append(a.get("href"))
    return all_href



url = "https://www.x-rates.com/table/?from=USD&amount=1"
for item in html_parsing(url):
    print(item)
```

```python
# 2. Напишите программу, которая запрашивает у пользователя URL-адрес веб-страницы и
# уровень заголовков, а затем использует библиотеку Beautiful Soup для парсинга HTML и
# извлекает заголовки нужного уровня (теги h1, h2, h3 и т.д.) с их текстом.
from bs4 import BeautifulSoup
import requests


def html_parsing(url: str, teg: str) -> list[str]:
    response = requests.get(url).text
    soup_response = BeautifulSoup(response, "html.parser")
    all_tegs = soup_response.find_all(teg)
    headers = []
    for header in all_tegs:
        headers.append(header.getText())
    return headers



url = "https://books.toscrape.com/"
for item in html_parsing(url, "h3"):
    print(item)
```
